```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string

import warnings
warnings.filterwarnings('ignore')
```

```python
real = pd.read_csv('True.csv.zip')
fake = pd.read_csv('Fake.csv.zip')
print(real.head())
print(fake.head())
```

```python
fake['Class'] = 0
real['Class'] = 1
```

```python
real.shape , fake.shape
```

```python
fake_manual_testing = fake.tail(10)
real_manual_testing = real.tail(10)

real.drop(real_manual_testing.index, inplace=True)
fake.drop(fake_manual_testing.index, inplace=True)
```

```python
real.shape , fake.shape
```

```python
fake_manual_testing['Class'] = 0
real_manual_testing['Class'] = 1
```

```python
fake_manual_testing.head(10)
```

```python
real_manual_testing.head(10)
```

```python
merged_data = pd.concat([fake , real ], axis=0)

merged_data.head(10)
```

```python
merged_data.columns
```

```python
#Removing the columns wich are not required .

data = merged_data.drop(['title' , 'subject' , 'date'] , axis=1)
```

```python
data.isnull().sum()
```

```python
#Random shuffeling of datasets

data = data.sample(frac=1)

data.head()

#Here, in th eoutput, 1 represents the true/real news dataset
# and 0 represents the fake news dataset.
```

```python
data.reset_index(inplace=True)
data.drop(['index'] , axis=1 , inplace=True)
```

```python
data.columns
```

```python
data.head()
```

```python
#Creating a function to process the text.

def word_check(text):
    text = text.lower()
    text = re.sub('\[.*?\]' , '' , text)
    text = re.sub("\\W" , " " , text)
    text = re.sub('https?://\S+|www\.\S+' , '' , text)
    text = re.sub('<.*?>+' , '' , text)
    text = re.sub('[%s]' % re.escape(string.punctuation) , '' , text)
    text = re.sub('\n' , '' , text)
    text = re.sub('\w*\d\w*' , '' , text)
    return text
```

```python
data['text'] = data['text'].apply(word_check)
```

```python
x = data['text']
y = data['Class']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.25)
```

```python
#Converting text to vector


from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train , y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test , y_test)
```

```python
print(classification_report(y_test , pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train , y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test , y_test)
```

```python
print(classification_report(y_test , pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state=0)
GB.fit(xv_train , y_train)
```

```python
pred_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test , y_test)
```

```python
print(classification_report(y_test , pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state=0)
RF.fit(xv_train , y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test , y_test)
```

```python
print(classification_report(y_test , pred_rf))
```

```python
#Creating output function

def output(n):
  if n == 0:
    return "Fake news"
  elif n == 1:
    return "Not a fake news"


def manual_testing(news):
  testing_news = {"text":[news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(word_check)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)
  pred_lr = LR.predict(new_xv_test)
  pred_dt = DT.predict(new_xv_test)
  pred_gb = GB.predict(new_xv_test)
  pred_rf = RF.predict(new_xv_test)
  return print("\n\nlR Prediction: {} \nDT Prediction: {} \nGB Prediction: {} \nRF Prediction: {}".format(output(pred_lr[0]) , output(pred_dt[0]) , output(pred_gb[0]) , output(pred_rf[0])))
```
